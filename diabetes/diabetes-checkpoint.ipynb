```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#Remove all outliers
lower = df["Glucose"].quantile(0.05)
upper = df["Glucose"].quantile(0.95)
df = df[(df["Glucose"] > lower) & (df["Glucose"] < upper)];
lower = df['BloodPressure'].quantile(0.05)
upper = df['BloodPressure'].quantile(0.95)
df = df[(df['BloodPressure'] > lower) & (df['BloodPressure'] < upper)];
lower = df['SkinThickness'].quantile(0.05)
upper = df['SkinThickness'].quantile(0.95)
df = df[(df['SkinThickness'] > lower) & (df['SkinThickness'] < upper)];
lower = df['BMI'].quantile(0.05)
upper = df['BMI'].quantile(0.95)
df = df[(df['BMI'] > lower) & (df['BMI'] < upper)];
lower = df['Age'].quantile(0.05)
upper = df['Age'].quantile(0.95)
df = df[(df['Age'] > lower) & (df['Age'] < upper)];
lower = df["Pregnancies"].quantile(0.05)
upper = df["Pregnancies"].quantile(0.95)
df = df[(df["Pregnancies"] > lower) & (df["Pregnancies"] < upper)];
```

```python
df.shape
```

```python
correlation = df.corr()
correlation
```

```python
sns.heatmap(correlation, annot = True, cmap = "Blues")
plt.show()
```

```python
sns.set_style('dark')
sns.regplot(x ='Age', y ='Pregnancies', data = df)
plt.show()
```
